## Univariate,Bivariate and MultiVariate Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')
```

```python
df.head()
```

```python
df.shape
```

### Univariate Analysis

```python
df["species"].unique()
```

```python
df_setosa=df.loc[df['species']=='setosa']
df_setosa
```

```python
df_virginica=df.loc[df['species']=='virginica']
df_versicolor=df.loc[df['species']=='versicolor']
```

```python
plt.plot(df_setosa['sepal_length'],np.zeros_like(df_setosa['sepal_length']),'^')      # x , y = 0
plt.plot(df_virginica['sepal_length'],np.zeros_like(df_virginica['sepal_length']),'^')
plt.plot(df_versicolor['sepal_length'],np.zeros_like(df_versicolor['sepal_length']),'^')
plt.xlabel('Sepal length')
plt.show()
```

```python
df_setosa['sepal_length']
```

```python
plt.plot(df_setosa['petal_length'],np.zeros_like(df_setosa['sepal_length']),'^')      # x , y = 0
plt.plot(df_virginica['petal_length'],np.zeros_like(df_virginica['sepal_length']),'^')
plt.plot(df_versicolor['petal_length'],np.zeros_like(df_versicolor['sepal_length']),'^')
plt.xlabel('petal length')
plt.show()
```

```python
df['species'].unique()
```

```python
df['species'].value_counts()
```

### Bivariate Analysis

```python
sns.FacetGrid(df,hue="species",size=5).map(plt.scatter,"petal_length","sepal_width").add_legend();
plt.show()
```

```python
sns.catplot("petal_length" , "sepal_width" , data = df , hue = "species")
```

### Multivariate Analysis

```python
sns.pairplot(df,hue="species",size=3)
```

```python

g = sns.PairGrid(df, hue="species")
g = g.map_diag(plt.hist)
g = g.map_offdiag(plt.scatter)
g = g.add_legend()
```
